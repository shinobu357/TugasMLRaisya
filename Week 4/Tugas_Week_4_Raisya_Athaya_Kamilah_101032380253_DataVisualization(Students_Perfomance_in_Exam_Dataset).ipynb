Beri sample kode python untuk dapat menampilkan persentase dari data bertipe kategorikal dalam bentuk pie chart, sehingga saya bisa copy paste ke google colab.

Bagian untuk membuat dan menampilkan visualisasi dalam bentuk pie chart berupa nilai persentase dari data bertipe kategorikal. Misal Pie chart pada kategori gender, Female 51,8% dan Male 48,2% yang berarti lebih banyak Female daripada Male pada dataset tersebut

````python
# Mengatur gaya seaborn
sns.set(style="whitegrid")

# Fungsi untuk membuat pie chart
def plot_pie_chart(column_name):
    data = df[column_name].value_counts()  # Menghitung jumlah unik setiap kategori
    labels = data.index  # Mendapatkan label dari nilai unik
    sizes = data.values  # Mendapatkan nilai untuk masing-masing kategori

    plt.figure(figsize=(7, 7))  # Ukuran plot
    colors = sns.color_palette("pastel", len(labels))  # Palet warna pastel
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)  # Membuat pie chart

    plt.axis('equal')  # Membuat pie chart bulat
    plt.title(f'Percentage of {column_name}', fontsize=14, fontweight='bold')  # Menambahkan judul
    plt.show()  # Menampilkan plot


# Menampilkan pie chart untuk kolom-kolom kategorikal
plot_pie_chart('gender')
plot_pie_chart('race/ethnicity')
plot_pie_chart('parental level of education')
plot_pie_chart('lunch')
plot_pie_chart('test preparation course')
````

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

Bagian untuk membuat dan menampilkan visualisasi dalam bentuk diagram batang untuk menghitung jumlah siswa yang lulus dan tidak lulus pada kategori(math,reading,writing) dengan nilai minimal 75 jika siswa lulus.

Pada hasil visual tersebut paling banyak siswa tidak lulus adalah di math score dengan jumlah 700 orang, sedangkan pada reading dan writing score yang tidak lulus sekitar 600-650 orang.

Kategori yang paling banyak siswa lulus adalah di reading score di sekitar 350 orang, sedangkan pada math score yang lulus sekitar 290 orang dan writing score yang lulus sekitar 340 orang.

````python
# Menentukan nilai ambang batas
pass_threshold = 75

# Menghitung jumlah siswa yang lulus dan gagal untuk setiap pelajaran
data = {
    'Subject': ['Math', 'Reading', 'Writing'],
    'Pass': [
        (df['math score'] >= pass_threshold).sum(),
        (df['reading score'] >= pass_threshold).sum(),
        (df['writing score'] >= pass_threshold).sum()
    ],
    'Fail': [
        (df['math score'] < pass_threshold).sum(),
        (df['reading score'] < pass_threshold).sum(),
        (df['writing score'] < pass_threshold).sum()
    ]
}

# Membuat DataFrame untuk visualisasi
results_df = pd.DataFrame(data)

# Mengubah format DataFrame untuk mempermudah visualisasi
results_df = results_df.melt(id_vars='Subject', var_name='Result', value_name='Count')

# Mengatur gaya seaborn
sns.set(style="whitegrid")

# Membuat plot batang
plt.figure(figsize=(10, 6))
sns.barplot(x='Result', y='Count', hue='Subject', data=results_df, palette='pastel')

# Menambahkan judul dan label
plt.title('Jumlah Siswa yang Lulus dan Gagal untuk Masing-Masing Pelajaran', fontsize=16)
plt.xlabel('Hasil', fontsize=12)
plt.ylabel('Jumlah Siswa', fontsize=12)
plt.legend(title='Pelajaran')

# Menampilkan plot
plt.show()
````

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

Berdasarkan hasil terminal dapat disimpulkan bahwa nilai rata-rata siswa sangat dipengaruhi oleh kemampuan mereka dalam math,reading,dan writing. Namun faktor faktor seperti pendidikan orang tua,ikut kursus persiapan ujian,gender, dan makan siang memang dapat berpengaruh atau berperan penting dalam nilai rata-rata siswa , tetapi tidak sepenuhnya faktor utama. Karena yang paling dominan adalah dari nilai utama math,reading dan writing

````python
# Membuat kolom baru 'average score' yang merupakan rata-rata dari math score, reading score, dan writing score
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Melihat 5 baris pertama data untuk memastikan kolom 'average score' sudah ditambahkan
print(df.head())

# Mengubah kolom kategori ke dalam bentuk numerik menggunakan teknik one-hot encoding atau label encoding
df_encoded = pd.get_dummies(df, columns=['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course'], drop_first=True)

# Melihat korelasi antara average score dengan kategori yang sudah di-encode
correlation_matrix = df_encoded.corr()

# Menampilkan korelasi antara 'average score' dengan kolom-kolom lainnya
correlation_with_average = correlation_matrix['average score'].sort_values(ascending=False)
print(correlation_with_average)
````

Membuat visualisasi untuk menampilkan jumlah siswa nilai rata-rata math,reading,dan writingnya berdasarkan range nilai:

A= 81-100

B= 61-80

C= 41-60

D= Kurang dari 40.

Berdasarkan hasil visualisasi jumlah siswa yang paling banyak adalah yang dapat nilai B

````python
# Hitung rata-rata nilai untuk setiap siswa
df['average_score'] = (df['math score'] + df['reading score'] + df['writing score']) / 3

# Tentukan kategori nilai berdasarkan rentang yang ditentukan
def categorize_score(score):
    if score >= 81:
        return 'A'
    elif score >= 61:
        return 'B'
    elif score >= 41:
        return 'C'
    else:
        return 'D'

df['grade'] = df['average_score'].apply(categorize_score)

# Buat plot untuk menunjukkan jumlah siswa di setiap kategori
plt.figure(figsize=(10, 6))

# Pastikan hanya menggunakan grade tanpa hue
palette = ['lightgreen', 'lightblue', 'lightcoral', 'lightgray']
sns.countplot(data=df, x='grade', palette=palette)

# Tambahkan label dan judul
plt.title('Jumlah Siswa Berdasarkan Kategori Nilai Rata-rata')
plt.xlabel('Kategori Nilai')
plt.ylabel('Jumlah Siswa')

# Tampilkan plot
plt.show()
````
